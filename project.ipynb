## Content Base Filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix_content = tfidf_vectorizer.fit_transform(df2['content'])
cosine_similarities_content = cosine_similarity(tfidf_matrix_content, tfidf_matrix_content)
def content_filtering_recommendation(item_idx, k=5):
    similar_items = list(enumerate(cosine_similarities_content[item_idx]))
    similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)
    top_similar_items = similar_items[1:k+1]
    
    indices = [x[0] for x in top_similar_items]
    return indices
```

```python
similar_items = content_filtering_recommendation(2)
similar_items
```

## Collabrative Filtering

```python
def collaborative_filtering_recommendation(target_user_id):
    user_similarity = cosine_similarity(user_item_matrix)
    target_user_index = user_item_matrix.index.get_loc(target_user_id)
    user_similarities = user_similarity[target_user_index]
    similar_user_indices = user_similarities.argsort()[::-1][1:]
    recommended_items = []
    for user_index in similar_user_indices:
        rated_by_similar_user = user_item_matrix.iloc[user_index]
        not_rated_by_target_user = (rated_by_similar_user == 0) & (user_item_matrix.iloc[target_user_index] == 0)
        recommended_items.extend(user_item_matrix.columns[not_rated_by_target_user][:10])
    recommended_items_details = df1[df1['product id'].isin(recommended_items)][['product id']]
    result = recommended_items_details['product id'].head(10).tolist()
    final_result = df2[df2['Uniqe Id'].isin(result)]
    return final_result
```

```python
similar_items = collaborative_filtering_recommendation(500)
similar_items
```

```python
def hybrid_recommendation(target_user_id, item):
    content_based = content_filtering_recommendation(item)

    collabrative_based = collaborative_filtering_recommendation(target_user_id)

    hybrid_rec = pd.concat([content_based, collabrative_based])

    return hybrid_rec
```

```python
similar_recommendation = hybrid_recommendation(4, 2)
similar_recommendation
```

## Evaluation Criteria

```python
def train_test_split_user_item(matrix, test_ratio=0.2, seed=42):
    np.random.seed(seed)

    train = matrix.copy()
    test = pd.DataFrame(0, index=matrix.index, columns=matrix.columns)

    for user in matrix.index:
        rated_items = matrix.columns[matrix.loc[user] > 0].tolist()

        if len(rated_items) < 2:
            continue

        n_test = max(1, int(len(rated_items) * test_ratio))
        test_items = np.random.choice(rated_items, size=n_test, replace=False)

        for item in test_items:
            test.loc[user, item] = matrix.loc[user, item] 
            train.loc[user, item] = 0                     

    return train, test
```

```python
def precision_recall_f1_content(item_idx, k=5):
    recommended_items = content_filtering_recommendation(item_idx, k)

    hit = 1 if item_idx in recommended_items else 0

    precision = hit / k
    recall = hit 

    if precision + recall == 0:
        f1 = 0
    else:
        f1 = 2 * precision * recall / (precision + recall)

    return precision, recall, f1
```

```python
P, R, F = [], [], []

for item_idx in range(len(df2)):
    p, r, f = precision_recall_f1_content(item_idx, k=5)
    P.append(p); R.append(r); F.append(f)

print("Precision@5:", sum(P)/len(P))
print("Recall@5:", sum(R)/len(R))
print("F1@5:", sum(F)/len(F))
```
