```python
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
from scipy.sparse.linalg import svds
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_interactions = pd.read_csv('users_interactions.csv')
df_products = pd.read_csv('products_catalog.csv')
```

```python
user_activity = df_interactions["user_id"].value_counts()

plt.figure()
plt.hist(user_activity, bins=30)
plt.title("User Interaction Distribution")
plt.xlabel("Number of Interactions per User")
plt.ylabel("Number of Users")
plt.show()
```

```python
user_item_matrix = df_interactions.pivot_table(
    index="user_id",
    columns="product_id",
    values="rating",
    aggfunc="mean"
).fillna(0)

product_ids   = user_item_matrix.columns
sparse_matrix = csr_matrix(user_item_matrix.values)

user_item_matrix
```

```python
# SVD MODEL
k = min(50, min(sparse_matrix.shape) - 1)

U, sigma, Vt = svds(sparse_matrix, k=k)

predicted_ratings = np.dot(np.dot(U, np.diag(sigma)), Vt)

print(f"[SVD] shape: {predicted_ratings.shape}")
```

```python
# ITEM-ITEM COSINE SIMILARITY
item_similarity_matrix = cosine_similarity(user_item_matrix.T)
item_similarity_df = pd.DataFrame(
    item_similarity_matrix,
    index=product_ids,
    columns=product_ids
)

item_similarity_df.head()
```

```python
plt.figure()
plt.imshow(item_similarity_matrix)
plt.title("Item-Item Similarity Matrix")
plt.xlabel("Products")
plt.ylabel("Products")
plt.show()
```

```python
# POPULARITY BASELINE
product_popularity = df_interactions["product_id"].value_counts()

existing_users = set(user_item_matrix.index.tolist())
```

```python
# UTILITY FUNCTIONS
def products_to_list(df_slice: pd.DataFrame) -> list:
    cols = ["product_id", "product_name", "brand", "category",
            "price", "description", "image_url"]
    return df_slice[cols].to_dict(orient="records")


def recommend_popular(top_n: int = 12) -> list:
    pids = product_popularity.head(top_n).index
    return products_to_list(
        df_products[df_products["product_id"].isin(pids)]
    )
```

```python
# SVD RECOMMENDATION
def recommend_svd(user_id: str, top_n: int = 12, exclude_pids: list = None) -> list:

    user_index = user_item_matrix.index.get_loc(user_id)
    user_preds = predicted_ratings[user_index].copy()

    already = sparse_matrix[user_index].toarray().flatten()
    user_preds[already > 0] = -np.inf

    if exclude_pids:
        for pid in exclude_pids:
            if pid in product_ids:
                user_preds[product_ids.get_loc(pid)] = -np.inf

    rec_indices = np.argsort(user_preds)[-top_n:][::-1]
    rec_pids    = product_ids[rec_indices]

    return products_to_list(
        df_products[df_products["product_id"].isin(rec_pids)]
    )
```

```python
# ITEM-BASED CF
def get_similar_for_product(pid: str, exclude_pids: list, top_k: int) -> list:

    if pid not in item_similarity_df.index or top_k <= 0:
        return []

    scores = item_similarity_df[pid].copy()
    scores = scores.drop(labels=exclude_pids, errors="ignore")

    return scores.sort_values(ascending=False).head(top_k).index.tolist()


def recommend_item_based(user_product_ids: list, top_n: int = 12) -> list:

    if not user_product_ids:
        return []

    slots       = get_slot_allocation(len(user_product_ids), top_n)
    all_rec_ids = []
    excluded    = list(user_product_ids)

    for pid, slot_size in zip(user_product_ids, slots):
        similar = get_similar_for_product(pid, excluded, slot_size)
        all_rec_ids.extend(similar)
        excluded.extend(similar)

    if not all_rec_ids:
        return []

    order_map = {pid: i for i, pid in enumerate(all_rec_ids)}

    rec_df = df_products[df_products["product_id"].isin(all_rec_ids)].copy()
    rec_df["_order"] = rec_df["product_id"].map(order_map)

    return products_to_list(
        rec_df.sort_values("_order").drop(columns="_order")
    )
```

```python
# USER-BASED CF
def recommend_user_based(user_product_ids: list, top_n: int = 12) -> list:

    if not user_product_ids:
        return recommend_popular(top_n)

    vec = np.zeros(user_item_matrix.shape[1])

    for pid in user_product_ids:
        if pid in user_item_matrix.columns:
            idx    = user_item_matrix.columns.get_loc(pid)
            mean_r = df_interactions[df_interactions["product_id"] == pid]["rating"].mean()
            vec[idx] = mean_r if not np.isnan(mean_r) else 3.0

    sims = cosine_similarity(
        vec.reshape(1, -1),
        user_item_matrix.values
    ).flatten()

    top_indices   = sims.argsort()[-10:][::-1]
    similar_users = user_item_matrix.index[top_indices]

    similar_data = df_interactions[
        df_interactions["user_id"].isin(similar_users)
    ]

    counts = similar_data["product_id"].value_counts()
    counts = counts.drop(labels=user_product_ids, errors="ignore")

    rec_pids = counts.head(top_n).index

    return products_to_list(
        df_products[df_products["product_id"].isin(rec_pids)]
    )
```
